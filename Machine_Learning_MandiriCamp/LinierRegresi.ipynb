# Simple Linear Regression

```python
import pandas as pd
```

```python
df1 = pd.read_csv('bensin.csv')
```

```python
df1
```

```python
df1.describe()
```

```python
import numpy as np
import sklearn.model_selection as ms
import matplotlib.pyplot as plt
```

```python
liter = df1[['Liter']]
kilometer = df1[['Kilometer']]
X_train, X_test, y_train, y_test = ms.train_test_split(liter, kilometer, test_size=0.2, random_state=0)
plt.scatter(X_train, y_train, edgecolors='r')
plt.xlabel('Liter')
plt.ylabel('Kilometer')
plt.title('Konsumsi Bahan Bakar')
x1 = np.linspace(0,45)
y1 = 4 + 7 * x1
plt.plot(x1,y1)
plt.show()
```

```python
import sklearn.linear_model as lm
model1 = lm.LinearRegression()
model1.fit(X_train, y_train)
```

```python
liter = df1[['Liter']]
kilometer = df1[['Kilometer']]
X_train, X_test, y_train, y_test = ms.train_test_split(liter, kilometer, test_size=0.2, random_state=0)
plt.scatter(X_train, y_train, edgecolors='r')
plt.xlabel('Liter')
plt.ylabel('Kilometer')
plt.title('Konsumsi Bahan Bakar')
x1=np.linspace(0,45)
y1=3.94+6.67*x1
plt.plot(x1,y1)
plt.show()
```

```python
r2 = model1.score(X_test, y_test)
print(r2)
```

```python
jarak = model1.predict([[60]])
print(jarak)
```
